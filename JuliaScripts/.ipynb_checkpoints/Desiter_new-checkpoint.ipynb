```julia
using Plots
using ColorSchemes
using QuadGK
using LaTeXStrings
using Dierckx
using Makie

energy1(R0) = R0*sqrt(1-R0^2)
energy2(R0) = R0*(1+R0^2)/sqrt(1-R0^2)
energy1(R0, H) = R0*sqrt(1 - H^2*R0^2)

function hamilton_new(R,E,sgn, H)
    (H^3*R.^5 - H*R.^3 + E^2*H*R .+ sgn * sqrt.(E^2*(E^2-R.^2+H^2*R.^4)))./(E^2 .+ H^2*R.^4)
end

R00(E, H, sgn) = sqrt(1.0 + sgn*sqrt(1.0-4.0*H^2*E^2))/(sqrt(2.0)*H)
```

```julia
gr(fmt = :png)
H = 1
step = 0.0001
R_lim = 1.3
EE = (2.0,1.2, 0.7, 0.5, 0.4, 0.25, 0.08)
R = -R_lim:3*step:R_lim

p1 = Plots.plot(R, R-1, seriestype = :path, label = "", c = :black, legend = :topleft)
Plots.plot!(R, R+1, seriestype = :path, label = "", c = :black, legend = :topleft)

for i = 1:size(EE,1)
    E = EE[i]
    if E < 0.5
        continue
    end
    R = -R_lim:3*step:R_lim
    Rt = hamilton_new(R, E, 1, H)
    Rtm = hamilton_new(R, E, -1, H)
    Plots.plot!(R, Rt, seriestype = :path, label = "E = $(round(E, 2))", c = ColorSchemes.Set1_7[i])
    Plots.plot!(R, Rtm, seriestype = :path, label = "", c = ColorSchemes.Set1_7[i])
end

for i = 1:size(EE,1)
    E = EE[i]
        if E >= 0.5
        continue
    end
    R0 = R00(E, H, -1)
    R1 = R00(E, H, 1)
    R_I = -R0+step:step:R0-step
    Rm_O = -R_lim:step:-R1-step
    R_O = R1+step:step:R_lim
    Rt_I = hamilton_new(R_I, E, 1, H)
    Rtm_I = hamilton_new(R_I, E, -1, H);
    Rt_O = [hamilton_new(R_O[end:-1:1], E, 1, H); hamilton_new(R_O, E, -1, H)]
    Rtm_O = [hamilton_new(Rm_O, E, 1, H); hamilton_new(Rm_O[end:-1:1], E, -1, H)]
    Rtt = [Rtm_I; Rt_I[end:-1:1]; Rtm_I[1]]
    RR = [R_I; R_I[end:-1:1]; R_I[1]]
    Plots.plot!(RR, Rtt, seriestype = :path, label = "E = $(round(E, 2))", c = ColorSchemes.Set1_7[i])
    Plots.plot!([R_O[end:-1:1]; R_O], Rt_O, c = ColorSchemes.Set1_7[i], label = "")
    Plots.plot!([Rm_O; Rm_O[end:-1:1]], Rtm_O, c = ColorSchemes.Set1_7[i], label = "")
end
Plots.plot(p1, size = (800, 600), linewidth = 2, xlabel = "R", ylabel = "dR/dt")
#savefig(p1, "DeSitter.pdf")
```

```julia
H = 1
N = 1000
EE = (0.5, 0.4, 0.25, 0.08)
result = Array{Any, 1}(size(EE,1))
resultm = Array{Any, 1}(size(EE,1))
RR = Array{Any, 1}(size(EE,1))
maxx = Array{Any, 1}(size(EE,1))

for i = 1:size(EE,1)
    E = EE[i]
    R0 = R00(E, H, -1)
    step = 2*R0/(N+1)
    R = R0-step:-step:-R0+step
    des_integral(R) = 1/hamilton_new(R, E, 1, H)
    des_integralm(R) = 1/hamilton_new(R, E, -1, H)
    
    RR[i] = R
    res = [quadgk(des_integral, R[1], R[i]) for i = 1:size(R,1)]
    resm = [quadgk(des_integralm, R[1], R[i]) for i = 1:size(R,1)]
    
    result[i] = res
    resultm[i] = resm
    maxx[i] = maximum(resm)
end
fin_res = [([[resultm[a][i][1] for i = 1:size(resultm[a],1)]; ([result[a][i][1] for i = 1:size(result[a],1)]+2*maxx[a][1])[end:-1:1]; [resultm[a][i][1] for i = 1:size(resultm[a],1)]+2*maxx[a][1]; ([result[a][i][1] for i = 1:size(result[a],1)]+4*maxx[a][1])[end:-1:1]], [RR[a]; RR[a][end:-1:1]; RR[a]; RR[a][end:-1:1]] ) for a = 1:size(result,1)]
```

```julia
p1 =  Plots.plot(fin_res[1], label = "E = $(EE[1])")

p2 =  Plots.plot(fin_res[2], label = "E = $(EE[2])")

p3 =  Plots.plot(fin_res[3], label = "E = $(EE[3])")

p4 =  Plots.plot(fin_res[4], label = "E = $(EE[4])", xlabel = "t")

pp = Plots.plot(p1, p2, p3, p4, layout = (4,1), size = (800, 600), linewidth = 2, ylabel = "R", legend = :topright)

#savefig(pp, "Time_dep.pdf")
```

```julia
t00 = 0.0
t_fin = 3*pi
r0 = 1.0
n = 0
N = 100
k = 0

scene = Scene(resolution = (1000, 1000))

f(t, s, r0, n) = (r0*cos(t/r0)*cos(n*s+k*pi/2)*cos(s), r0*cos(t/r0)*cos(n*s+k*pi/2)*sin(s))

function surf_func(t0, t_fin, N, n, r0)
    t_range = range(t0, stop = t_fin+t0, length = N)
    s_range = range(0, stop = 2pi, length = N)
    srf = [f(t, s, r0, n) for t in t_range, s in s_range]
    x = [srf[i,j][1] for i = 1:N, j = 1:N]
    y = [srf[i,j][2] for i = 1:N, j = 1:N]
    z = [t-t_fin for t in t_range, s in s_range]
    return x, y, z
end

x, y, z = surf_func(t00, t_fin, N, n, r0)

surf = surface!(scene, z, x, y, colormap = :Blues)[end]
str = lines!(scene, z[end,:], x[end,:], y[end,:], color = :red, linewidth = 2, limits = FRect3D(Vec3f0(-10, -1.1, -1.1), Vec3f0(20, 2.2, 2.2)))[end]
update_cam!(scene, Vec3f0(4, 8, 8), Vec3f0(-3, 0, 0))

record(scene, "D:\\Documents\\Diplomka\\string_surf"*"$n"*".mp4", range(t00, stop = t_fin, length = 300)) do i
    x, y, z = surf_func(i, t_fin, N, n, r0)
    surf[1] = z
    surf[2] = x
    surf[3] = y
    str[1] = z[end,:]
    str[2] = x[end,:]
    str[3] = y[end,:]
    update_cam!(scene, Vec3f0(4+i, 8, 8), Vec3f0(-3+i, 0, 0))
end
```
