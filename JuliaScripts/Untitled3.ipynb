```julia
using Plots
using Unitful
using AxisArrays
```

```julia
global const h = 6.62606896e-34
global const c = 3.0e8
global const kB = 1.3806504e-23
global const em = 1.602176487e-19
T0 = 2.725
```

```julia
function planck(x, T)
    2*h/c^2*x.^3./(exp.(h*x/kB*T)-1)
end    
```

```julia
xmax = 1e11
step = xmax/100000
xdata = step:step:xmax
ydata = planck(xdata,T0)
result = AxisArray(ydata, Axis{:time}(xdata.*u"Hz"));
```

```julia
@recipe function f(data::AxisArray; units = u"kHz", yticks = :auto)
    seriestype  :=  :path
    xlabel := string(units)
    legend --> :none
    #formatter := :scientific
    
    xdata = Float64.(axes(data)[1][1:end]./units)
    ydata = data.data
    
    yticks := yticks
    
    
    xdata, ydata
end
```

```julia
yticks = ([0, 2e-21, 4e-21, 6e-21, 8e-21, 10e-21],[0.0, 2e-21, 4e-21, 6e-21, 8e-21, 10e-21])
plot(result, units = u"GHz", yticks = yticks)
ylims!(0,10e-21)
```
