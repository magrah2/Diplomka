```julia
using Plots
gr(fmt = :png)
pyplot(dpi = 300)
using LaTeXStrings
using Measures
```

```julia
meshgrid(x, y) = (repeat(x, outer=length(y)), repeat(y, inner=length(x)))
x, y = meshgrid(-0.1:0.02:0.1, -0.1:0.02:0.1)
H = 1.0
scale = 0.07
u = @. scale*y
v = @. scale*(3*H*y - 2*H^2*x)
p1 = plot()

s1 = 0.12
x1 = [0.0, s1]
y1 = [0.0, s1*H]
plot!(x1, y1, arrow = arrow(1.5, 1), linecolor = :red, linewidth = :1.5)
x2 = [0.0, -s1]
y2 = [0.0, -s1*H]
plot!(x2, y2, arrow = arrow(1.5, 1), linecolor = :red, linewidth = :1.5)

s2 = 0.07
x3 = [0.0, s2]
y3 = [0.0, s2*2*H]
plot!(x3, y3, arrow = arrow(1.5, 1), linecolor = :red, linewidth = :1.5)
x4 = [0.0, -s2]
y4 = [0.0, -s2*2*H]
plot!(x4, y4, arrow = arrow(1.5, 1), linecolor = :red, 
    xlabel = L"^{^{^{^.}}} \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad R",
    ylabel = L"^{^{^{^.}}} \qquad \qquad \quad \qquad \qquad \qquad \qquad \qquad \qquad \qquad S",
    legend = :none, linewidth = 2, tickfontsize = 15,
    xticks = ([0], [L"\frac{1}{H}"]), yticks = ([0], [L"0"]), size = (800, 600), guidefontsize = 18, gridalpha = 0.2, gridlinewidth = 1.0)
quiver!(x, y, quiver=(u, v), linewidth = 1.0, c=:blue)
savefig(p1, "unstable_node.pdf")
```

```julia
meshgrid(x, y) = (repeat(x, outer=length(y)), repeat(y, inner=length(x)))
x, y = meshgrid(-0.1*sqrt(2):0.02*sqrt(2):0.1*sqrt(2), -0.2:0.04:0.2)
H = 1.0
scale = 0.1
u = @. scale*y
v = @. scale*2*H^2*x
p2 = plot()

s1 = 0.17
x1 = [0.0, s1]
y1 = [0.0, s1*sqrt(2)*H]
plot!(x1, y1, arrow = arrow(1.5, 1), linecolor = :red, linewidth = 2.0)
x2 = [0.0, -s1]
y2 = [0.0, -s1*sqrt(2)*H]
plot!(x2, y2, arrow = arrow(1.5, 1), linecolor = :red, linewidth = 2.0)

s2 = 0.17
x3 = [s2, 0.0]
y3 = [-s2*sqrt(2)*H, 0.0]
plot!(x3, y3, arrow = arrow(1.5, 1), linecolor = :red, linewidth = 2.0)
x4 = [-s2, 0.0]
y4 = [s2*sqrt(2)*H, 0.0]
plot!(x4, y4, arrow = arrow(1.5, 1), linecolor = :red, 
    xlabel = L"_{_{_{_.}}}  \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad R",
    ylabel = L"^{^{^{^.}}} \qquad \qquad \quad \qquad \qquad \qquad \qquad \qquad \qquad \qquad S",
    legend = :none, linewidth = 2, tickfontsize = 15,
    xticks = ([0], [L"\frac{1}{\sqrt{2}H}"]), yticks = ([0], ["0"]), size = (800, 600), guidefontsize = 18, gridalpha = 0.2, gridlinewidth = 1.0)
quiver!(x, y, quiver=(u, v), linewidth = 1.0, c=:blue)
savefig(p2, "saddle.pdf")
```
