```julia
using LsqFit
using Plots
```

```julia
model(x, p) = p[1]*exp.(-p[2]./x) + p[3]
```

```julia
xdata = [1251.51143893183
1270.91269918723
1298.74503757155
1330.10474996276
1353.61359885974
1376.87895062573
1400.28982065812
1416.78349576952]
```

```julia
ydata = [1.00824859048325e-09
9.82319250196874e-10
1.06531690358372e-09
1.12541906909351e-09
1.23372410421220e-09
1.38471133877356e-09
1.44431301166467e-09
1.55410225905179e-09]
```

```julia
xdat = deepcopy(xdata)
for i = 1:size(xdata,1)
    xdat[i] = xdata[size(xdata,1)+1-i]
end
xdata = xdat
```

```julia
x = 1200:10:1400
ydat = model(x, [-1e-8, 1.6/8.5e-3, 0])
plot(x,ydat)
```

```julia
fit = curve_fit(model, xdata, ydata, [5e-4,1.6/8.5e-5,8e-10])
```

```julia
fit.param
```

```julia
ydat = model(xdata,fit.param)

scatter(xdata,ydata)
plot!(xdata, ydat)
```

```julia
plot(xdata, model(xdata, [5e-4,1.6/8.5e-5,8e-10]))
scatter!(xdata,ydata)
```
