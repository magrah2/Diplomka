```julia
using Plots
pyplot()
using LaTeXStrings
```

```julia
A = 1
sigma = range(0.0, stop = 2*pi, length = 500)
tau = range(0.0, stop = pi, length = 100)
x = [[A*cos(s)*cos(t) for s in sigma] for t in tau];
y = [[A*sin(s)*cos(t) for s in sigma] for t in tau];
```

```julia
anim = @animate for t = 1:size(tau,1)
    p1 = plot(x[t], y[t], legend = :none, aspect_ratio = 1, linewidth = 2.5, xlabel = L"x", ylabel = L"y",
    xlims = (-1.1, 1.1), ylims = (-1.1, 1.1), legendfontsize = 15, guidefontsize = 20, tickfontsize = 12, size = (600, 600))
    savefig(p1, "D:/Documents/Diplomka/Animations/string_flat/string_flat_$(t).pdf")
end
```
