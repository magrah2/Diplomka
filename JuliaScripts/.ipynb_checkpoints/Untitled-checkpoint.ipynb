```python
from sage.calculus.desolvers import desolve_odeint
```

```python
x,y = var('x,y')
```

```python
f = [y, x+(E*x - sqrt(x^2*(x^2-1)+E^2))/sqrt(E^2*x^2+x^2*(x^2-1)+E^2 - 2*E*x*sqrt(x^2*(x^2-1)+E^2)+x^2*(x^2-1))]
```

```python
E = 10
```

```python
sol = desolve_odeint(f, [0.5,0], srange(0,3,0.0001), [x,y])
p=line(zip(sol[:,0],sol[:,1]))
g= plot(p)
g.show()
```

```python
g = Graphics()
for i in range(1,10):
    sol = desolve_odeint(f, [0.1+i*0.05,0], srange(0,3,0.0001), [x,y])
    p=line(zip(sol[:,0],sol[:,1]))
    g+= plot(p)
```

```python
g.show()
```
